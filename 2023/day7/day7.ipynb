```python
f = open("input_data.txt", "r")
data = f.read().split("\n")
```

```python
cards = [
    "A", 
    "K", 
    "Q", 
    "J", 
    "T", 
    "9", 
    "8", 
    "7", 
    "6", 
    "5", 
    "4", 
    "3",
    "2"
]

hand_results = [
    "High Card",
    "One Pair",
    "Two Pair",
    "Three of a Kind",
    "Full House",
    "Four of a Kind",
    "Five of a Kind"
    ]


card_strength = {k:v for k,v in zip(cards, list(range(len(cards)))[::-1])}
hand_strength = {k:v for k,v in zip(hand_results, list(range(len(hand_results)))[::-1])}

card_strength_str_vals = [
    "00",
    "01",
    "02",
    "03", 
    "04",
    "05",
    "06",
    "07",
    "08",
    "09",
    "10",
    "11",
    "12"
]

card_strength_str = {k:v for k,v in zip(cards, card_strength_str_vals[::-1])}


hands = []
bids = []
for x in data:
    hand, bid = x.split(" ")[0], x.split(" ")[1]
    hands.append(hand)
    bids.append(bid)

hand_bids_dict = {k:int(v) for k,v in zip(hands, bids)}
```

```python
def classify_hand(hand: str) -> str:
    """
    TBD
    """
    c_freq_dict = {card: hand.count(card) for card in hand}
    c_freq_list = c_freq_dict.values()
    c_freq_set = set(c_freq_list)
    
    max_freq = max(c_freq_set)
    min_freq = min(c_freq_set)

    if max_freq == 1:
        hand_classification = "High Card"
    elif max_freq == 2:
        count_of_pairs = len([x for x in c_freq_list if x == 2])
        if count_of_pairs == 1:
            hand_classification = "One Pair"
        elif count_of_pairs == 2:
            hand_classification = "Two Pair"
    elif max_freq == 3:
        if min_freq == 2:
            hand_classification = "Full House"
        elif min_freq == 1:
            hand_classification = "Three of a Kind"        
    elif max_freq == 4:
        hand_classification = "Four of a Kind"
    elif max_freq == 5:
        hand_classification = "Five of a Kind"

    return hand_classification


def assign_hands_to_dict(hands: list) -> dict:
    result_dict = {k: [] for k in hand_results}
    for h in hands:
        r = classify_hand(h)
        if result_dict[r] == []:
            result_dict[r] = [h]
        else:
            result_dict[r] = result_dict[r] + [h]

    return result_dict


def sort_ties(list_of_hands: list) -> list:
    for i in range(5):
        if i == 0:
            converted_to_vals = [f"{card_strength_str[hand[i]]}.0" for hand in list_of_hands]
        else:
            converted_to_vals = [x + y for x,y in zip(converted_to_vals, [f"0{card_strength_str[hand[i]]}" for hand in list_of_hands])]
    
    vals_dict = {k:float(v) for k,v in zip(list_of_hands, converted_to_vals)}
    return [x[0] for x in sorted(vals_dict.items(), key=lambda x:x[1])]


def create_final_sorted_list(list_of_hands: list) -> list:
    assigned_hands = assign_hands_to_dict(list_of_hands)

    for outcome in assigned_hands:
        if assigned_hands[outcome] != []:
            assigned_hands[outcome] = sort_ties(assigned_hands[outcome])

    return [value for key, values in assigned_hands.items() for value in values]


def solve_part_1(list_of_hands: list) -> int:
    final_sorted_list = create_final_sorted_list(list_of_hands)
    hands_ranking = {k:v + 1 for k,v in zip(final_sorted_list, list(range(len(final_sorted_list))))}

    return sum([hands_ranking[h] * hand_bids_dict[h] for h in hands])
```

```python
solve_part_1(hands)
```
